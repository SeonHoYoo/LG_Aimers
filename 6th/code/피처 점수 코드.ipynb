```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# ------------------------------
# 1. 초기 설정: 시술점수 칼럼 생성
# ------------------------------
# train_data에 시술점수 칼럼을 먼저 NaN으로 초기화
train_data['시술점수'] = np.nan

# ------------------------------
# 2. DI인 경우 처리
# ------------------------------
di_mask = train_data['시술 유형'] == 'DI'
if di_mask.sum() > 0:
    # 임신 시도 또는 마지막 임신 경과 연수 결측치는 0으로 채우기
    train_data.loc[di_mask, '임신시도_filled'] = train_data.loc[di_mask, '임신 시도 또는 마지막 임신 경과 연수'].fillna(0)
    
    # 클리닉 내 총 시술 횟수 및 DI 시술 횟수 매핑 (문자열 -> 숫자)
    procedure_mapping = {
        "0회": 0,
        "1회": 1,
        "2회": 2,
        "3회": 3,
        "4회": 4,
        "5회": 5,
        "6회 이상": 6
    }
    train_data.loc[di_mask, '클리닉시술_num'] = train_data.loc[di_mask, '클리닉 내 총 시술 횟수'].map(procedure_mapping)
    train_data.loc[di_mask, '불임원인'] = train_data.loc[di_mask, '부부 주 불임 원인'].astype(int)
    train_data.loc[di_mask, 'DI 시술'] = train_data.loc[di_mask, 'DI 시술 횟수'].map(procedure_mapping)
    
    # 학습에 사용할 피처와 타겟 설정
    features_DI = ['임신시도_filled', '클리닉시술_num', '불임원인', 'DI 시술']
    target = '임신 성공 여부'
    X_DI = train_data.loc[di_mask, features_DI]
    y_DI = train_data.loc[di_mask, target]
    
    # RandomForestRegressor 모델 학습 및 예측
    rf_model_di = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model_di.fit(X_DI, y_DI)
    pred_rf_di = rf_model_di.predict(X_DI)
    
    # 예측값을 시술점수 칼럼에 채워넣음
    train_data.loc[di_mask, '시술점수'] = pred_rf_di

# ------------------------------
# 3. IVF인 경우 처리
# ------------------------------
# IVF인 경우에는 시술유형 값을 아래와 같은 매핑 사전으로 0, 1, 2로 나눕니다.
ivf_mapping_dict = {
    'ICSI': 0,
    'IVF': 1,
    'Unknown': 2,
    'ICSI:ICSI': 0,
    'ICSI/BLASTOCYST': 0,
    'IVF / BLASTOCYST': 1,
    'IVF:IVF': 1,
    'ICSI:IVF': 0,
    'ICSI / AH': 0,
    'IVF:ICSI': 1,
    'IVF / AH': 1,
    'ICSI:Unknown': 0,
    'IVF:Unknown': 1,
    'ICSI / BLASTOCYST:IVF / BLASTOCYST': 0,
    'ICSI / BLASTOCYST :IVF / BLASTOCYST': 0,
    'IVF / AH:ICSI / AH': 1,
    'ICSI / AH:Unknown': 2,
    'ICSI / BLASTOCYST :ICSI': 2,
    'FER': 2,
    'GIFT': 2,
    'ICSI / BLASTOCYST ': 0
}

ivf_mask = train_data['시술 유형'] == 'IVF'
if ivf_mask.sum() > 0:
    # 시술유형매핑 칼럼 생성: 해당 행의 시술유형을 매핑 사전을 사용해 숫자로 변환
    train_data.loc[ivf_mask, '시술유형매핑'] = train_data.loc[ivf_mask, '시술 유형'].map(ivf_mapping_dict)
    
    # --- IVF 분기 내부에서 3가지 케이스로 나눔 ---
    # (1) 케이스 0: 시술유형매핑 값이 0인 경우 (ICSI 관련)
    ivf_case0_mask = ivf_mask & (train_data['시술유형매핑'] == 0)
    if ivf_case0_mask.sum() > 0:
        # 배아 이식 경과일 결측치는 0으로 채움
        train_data.loc[ivf_case0_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case0_mask, '배아 이식 경과일'].fillna(0)
        icsi_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled', 
            '미세주입된 난자 수', '미세주입에서 생성된 배아 수', 
            '미세주입 배아 이식 수', '미세주입 후 저장된 배아 수', '해동된 배아 수'
        ]
        X_icsi = train_data.loc[ivf_case0_mask, icsi_features]
        y_icsi = train_data.loc[ivf_case0_mask, target]
        
        rf_model_icsi = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_icsi.fit(X_icsi, y_icsi)
        pred_rf_icsi = rf_model_icsi.predict(X_icsi)
        train_data.loc[ivf_case0_mask, '시술점수'] = pred_rf_icsi
    
    # (2) 케이스 1: 시술유형매핑 값이 1인 경우 (IVF 관련)
    ivf_case1_mask = ivf_mask & (train_data['시술유형매핑'] == 1)
    if ivf_case1_mask.sum() > 0:
        train_data.loc[ivf_case1_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case1_mask, '배아 이식 경과일'].fillna(0)
        ivf_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled', 
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수', 
            '이식된 배아 수', '미세주입에서 생성된 배아 수', '미세주입된 난자 수', '해동된 배아 수'
        ]
        X_ivf = train_data.loc[ivf_case1_mask, ivf_features]
        y_ivf = train_data.loc[ivf_case1_mask, target]
        
        rf_model_ivf = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_ivf.fit(X_ivf, y_ivf)
        pred_rf_ivf = rf_model_ivf.predict(X_ivf)
        train_data.loc[ivf_case1_mask, '시술점수'] = pred_rf_ivf
    
    # (3) 케이스 2: 시술유형매핑 값이 2인 경우 (Unknown 관련)
    ivf_case2_mask = ivf_mask & (train_data['시술유형매핑'] == 2)
    if ivf_case2_mask.sum() > 0:
        train_data.loc[ivf_case2_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case2_mask, '배아 이식 경과일'].fillna(0)
        unknown_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled', 
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수', 
            '이식된 배아 수', '해동된 배아 수'
        ]
        X_unknown = train_data.loc[ivf_case2_mask, unknown_features]
        y_unknown = train_data.loc[ivf_case2_mask, target]
        
        rf_model_unknown = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_unknown.fit(X_unknown, y_unknown)
        pred_rf_unknown = rf_model_unknown.predict(X_unknown)
        train_data.loc[ivf_case2_mask, '시술점수'] = pred_rf_unknown

# ------------------------------
# 최종: train_data의 시술점수 칼럼이 채워졌습니다.
# ------------------------------
print("시술점수 칼럼 생성 완료")
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# ================================
# 0. 기본 설정 및 매핑 사전 정의
# ================================
# (실제 데이터 로딩: train_data, test_data가 이미 DataFrame으로 준비되어 있다고 가정)

# DI용 시술 횟수 매핑 사전
procedure_mapping = {
    "0회": 0,
    "1회": 1,
    "2회": 2,
    "3회": 3,
    "4회": 4,
    "5회": 5,
    "6회 이상": 6
}

# IVF용 상세 시술 유형 매핑 사전 (예: 시술 상세 칼럼에 있는 값 기준)
ivf_mapping_dict = {
    'ICSI': 0,
    'IVF': 1,
    'Unknown': 2,
    'ICSI:ICSI': 0,
    'ICSI/BLASTOCYST': 0,
    'IVF / BLASTOCYST': 1,
    'IVF:IVF': 1,
    'ICSI:IVF': 0,
    'ICSI / AH': 0,
    'IVF:ICSI': 1,
    'IVF / AH': 1,
    'ICSI:Unknown': 0,
    'IVF:Unknown': 1,
    'ICSI / BLASTOCYST:IVF / BLASTOCYST': 0,
    'ICSI / BLASTOCYST :IVF / BLASTOCYST': 0,
    'IVF / AH:ICSI / AH': 1,
    'ICSI / AH:Unknown': 2,
    'ICSI / BLASTOCYST :ICSI': 2,
    'FER': 2,
    'GIFT': 2,
    'ICSI / BLASTOCYST ': 0,
    'IVF:Unknown:Unknown:Unknown' : 1
}

# -------------------------------
# 1. TRAIN_DATA 처리 및 모델 학습
# -------------------------------
# train_data에 시술점수 칼럼 초기화
train_data['시술점수'] = np.nan
target = '임신 성공 여부'  # 타겟 칼럼

# ----- (1) DI 처리 -----
di_mask = train_data['시술 유형'] == 'DI'
if di_mask.sum() > 0:
    # 전처리: 결측치는 0으로 채우기
    train_data.loc[di_mask, '임신시도_filled'] = train_data.loc[di_mask, '임신 시도 또는 마지막 임신 경과 연수'].fillna(0)
    train_data.loc[di_mask, '클리닉시술_num'] = train_data.loc[di_mask, '클리닉 내 총 시술 횟수'].map(procedure_mapping)
    train_data.loc[di_mask, '불임원인'] = train_data.loc[di_mask, '부부 주 불임 원인'].astype(int)
    train_data.loc[di_mask, 'DI 시술'] = train_data.loc[di_mask, 'DI 시술 횟수'].map(procedure_mapping)
    
    features_DI = ['임신시도_filled', '클리닉시술_num', '불임원인', 'DI 시술']
    X_DI = train_data.loc[di_mask, features_DI]
    y_DI = train_data.loc[di_mask, target]
    
    # DI 모델 학습
    rf_model_di = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model_di.fit(X_DI, y_DI)
    pred_rf_di = rf_model_di.predict(X_DI)
    
    train_data.loc[di_mask, '시술점수'] = pred_rf_di

# ----- (2) IVF 처리 -----
ivf_mask = train_data['시술 유형'] == 'IVF'
if ivf_mask.sum() > 0:
    # IVF 행에서는 상세 시술 유형 칼럼("시술 상세")을 기준으로 매핑
    train_data.loc[ivf_mask, '시술유형매핑'] = train_data.loc[ivf_mask, '특정 시술 유형'].map(ivf_mapping_dict)
    
    # ● IVF Case 0 : 시술유형매핑 == 0 (예: ICSI 관련)
    ivf_case0_mask = ivf_mask & (train_data['시술유형매핑'] == 0)
    if ivf_case0_mask.sum() > 0:
        train_data.loc[ivf_case0_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case0_mask, '배아 이식 경과일'].fillna(0)
        icsi_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '미세주입된 난자 수', '미세주입에서 생성된 배아 수',
            '미세주입 배아 이식 수', '미세주입 후 저장된 배아 수', '해동된 배아 수'
        ]
        X_icsi = train_data.loc[ivf_case0_mask, icsi_features]
        y_icsi = train_data.loc[ivf_case0_mask, target]
        
        rf_model_icsi = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_icsi.fit(X_icsi, y_icsi)
        pred_rf_icsi = rf_model_icsi.predict(X_icsi)
        train_data.loc[ivf_case0_mask, '시술점수'] = pred_rf_icsi

    # ● IVF Case 1 : 시술유형매핑 == 1 (예: IVF 관련)
    ivf_case1_mask = ivf_mask & (train_data['시술유형매핑'] == 1)
    if ivf_case1_mask.sum() > 0:
        train_data.loc[ivf_case1_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case1_mask, '배아 이식 경과일'].fillna(0)
        ivf_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수',
            '이식된 배아 수', '미세주입에서 생성된 배아 수', '미세주입된 난자 수', '해동된 배아 수'
        ]
        X_ivf = train_data.loc[ivf_case1_mask, ivf_features]
        y_ivf = train_data.loc[ivf_case1_mask, target]
        
        rf_model_ivf = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_ivf.fit(X_ivf, y_ivf)
        pred_rf_ivf = rf_model_ivf.predict(X_ivf)
        train_data.loc[ivf_case1_mask, '시술점수'] = pred_rf_ivf

    # ● IVF Case 2 : 시술유형매핑 == 2 (예: Unknown 관련)
    ivf_case2_mask = ivf_mask & (train_data['시술유형매핑'] == 2)
    if ivf_case2_mask.sum() > 0:
        train_data.loc[ivf_case2_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case2_mask, '배아 이식 경과일'].fillna(0)
        unknown_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수',
            '이식된 배아 수', '해동된 배아 수'
        ]
        X_unknown = train_data.loc[ivf_case2_mask, unknown_features]
        y_unknown = train_data.loc[ivf_case2_mask, target]
        
        rf_model_unknown = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_unknown.fit(X_unknown, y_unknown)
        pred_rf_unknown = rf_model_unknown.predict(X_unknown)
        train_data.loc[ivf_case2_mask, '시술점수'] = pred_rf_unknown

print("Train_data에 시술점수 칼럼 추가 완료.")

# ================================
# 2. TEST_DATA 처리 및 예측 (학습한 모델 재사용)
# ================================
test_data['시술점수'] = np.nan

# ----- (1) DI 처리 (test_data) -----
di_mask_test = test_data['시술 유형'] == 'DI'
if di_mask_test.sum() > 0:
    test_data.loc[di_mask_test, '임신시도_filled'] = test_data.loc[di_mask_test, '임신 시도 또는 마지막 임신 경과 연수'].fillna(0)
    test_data.loc[di_mask_test, '클리닉시술_num'] = test_data.loc[di_mask_test, '클리닉 내 총 시술 횟수'].map(procedure_mapping)
    test_data.loc[di_mask_test, '불임원인'] = test_data.loc[di_mask_test, '부부 주 불임 원인'].astype(int)
    test_data.loc[di_mask_test, 'DI 시술'] = test_data.loc[di_mask_test, 'DI 시술 횟수'].map(procedure_mapping)
    
    X_DI_test = test_data.loc[di_mask_test, features_DI]
    pred_rf_di_test = rf_model_di.predict(X_DI_test)
    test_data.loc[di_mask_test, '시술점수'] = pred_rf_di_test

# ----- (2) IVF 처리 (test_data) -----
ivf_mask_test = test_data['시술 유형'] == 'IVF'
if ivf_mask_test.sum() > 0:
    # 상세 시술 유형("시술 상세") 기준으로 매핑
    test_data.loc[ivf_mask_test, '시술유형매핑'] = test_data.loc[ivf_mask_test, '특정 시술 유형'].map(ivf_mapping_dict)
    
    # ● IVF Case 0 (시술유형매핑 == 0)
    ivf_case0_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 0)
    if ivf_case0_mask_test.sum() > 0:
        test_data.loc[ivf_case0_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case0_mask_test, '배아 이식 경과일'].fillna(0)
        X_icsi_test = test_data.loc[ivf_case0_mask_test, icsi_features]
        pred_rf_icsi_test = rf_model_icsi.predict(X_icsi_test)
        test_data.loc[ivf_case0_mask_test, '시술점수'] = pred_rf_icsi_test

    # ● IVF Case 1 (시술유형매핑 == 1)
    ivf_case1_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 1)
    if ivf_case1_mask_test.sum() > 0:
        test_data.loc[ivf_case1_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case1_mask_test, '배아 이식 경과일'].fillna(0)
        X_ivf_test = test_data.loc[ivf_case1_mask_test, ivf_features]
        pred_rf_ivf_test = rf_model_ivf.predict(X_ivf_test)
        test_data.loc[ivf_case1_mask_test, '시술점수'] = pred_rf_ivf_test

    # ● IVF Case 2 (시술유형매핑 == 2)
    ivf_case2_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 2)
    if ivf_case2_mask_test.sum() > 0:
        test_data.loc[ivf_case2_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case2_mask_test, '배아 이식 경과일'].fillna(0)
        X_unknown_test = test_data.loc[ivf_case2_mask_test, unknown_features]
        pred_rf_unknown_test = rf_model_unknown.predict(X_unknown_test)
        test_data.loc[ivf_case2_mask_test, '시술점수'] = pred_rf_unknown_test

print("Test_data에 시술점수 칼럼 추가 완료.")
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# ================================
# 0. 기본 설정 및 매핑 사전 정의
# ================================
# (실제 데이터 로딩: train_data, test_data가 이미 DataFrame으로 준비되어 있다고 가정)

# DI용 시술 횟수 매핑 사전
procedure_mapping = {
    "0회": 0,
    "1회": 1,
    "2회": 2,
    "3회": 3,
    "4회": 4,
    "5회": 5,
    "6회 이상": 6
}

# IVF용 상세 시술 유형 매핑 사전 (예: 시술 상세 칼럼에 있는 값 기준)
ivf_mapping_dict = {
    'ICSI': 0,
    'IVF': 1,
    'Unknown': 2,
    'ICSI:ICSI': 0,
    'ICSI/BLASTOCYST': 0,
    'IVF / BLASTOCYST': 1,
    'IVF:IVF': 1,
    'ICSI:IVF': 0,
    'ICSI / AH': 0,
    'IVF:ICSI': 1,
    'IVF / AH': 1,
    'ICSI:Unknown': 0,
    'IVF:Unknown': 1,
    'ICSI / BLASTOCYST:IVF / BLASTOCYST': 0,
    'ICSI / BLASTOCYST :IVF / BLASTOCYST': 0,
    'IVF / AH:ICSI / AH': 1,
    'ICSI / AH:Unknown': 2,
    'ICSI / BLASTOCYST :ICSI': 2,
    'FER': 2,
    'GIFT': 2,
    'ICSI / BLASTOCYST ': 0,
    'IVF:Unknown:Unknown:Unknown': 1
}

# -------------------------------
# 1. TRAIN_DATA 처리 및 모델 학습
# -------------------------------
# train_data에 시술점수 칼럼 초기화
train_data['시술점수'] = np.nan
target = '임신 성공 여부'  # 타겟 칼럼

# ----- (1) DI 처리 -----
di_mask = train_data['시술 유형'] == 'DI'
if di_mask.sum() > 0:
    # 전처리: 결측치는 0으로 채우기
    train_data.loc[di_mask, '임신시도_filled'] = train_data.loc[di_mask, '임신 시도 또는 마지막 임신 경과 연수'].fillna(0)
    train_data.loc[di_mask, '클리닉시술_num'] = train_data.loc[di_mask, '클리닉 내 총 시술 횟수'].map(procedure_mapping)
    train_data.loc[di_mask, '불임원인'] = train_data.loc[di_mask, '부부 주 불임 원인'].astype(int)
    train_data.loc[di_mask, 'DI 시술'] = train_data.loc[di_mask, 'DI 시술 횟수'].map(procedure_mapping)
    
    features_DI = ['임신시도_filled', '클리닉시술_num', '불임원인', 'DI 시술']
    X_DI = train_data.loc[di_mask, features_DI]
    y_DI = train_data.loc[di_mask, target]
    
    # DI 모델 학습
    rf_model_di = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model_di.fit(X_DI, y_DI)
    pred_rf_di = rf_model_di.predict(X_DI)
    
    train_data.loc[di_mask, '시술점수'] = pred_rf_di

# ----- (2) IVF 처리 -----
ivf_mask = train_data['시술 유형'] == 'IVF'
if ivf_mask.sum() > 0:
    # IVF 행에서는 "특정 시술 유형" 칼럼의 결측값은 'IVF:Unknown:Unknown:Unknown'으로 채워서 매핑 (비어있으면 1로 지정)
    train_data.loc[ivf_mask, '시술유형매핑'] = train_data.loc[ivf_mask, '특정 시술 유형']\
        .fillna('IVF:Unknown:Unknown:Unknown')\
        .map(ivf_mapping_dict)
    
    # ● IVF Case 0 : 시술유형매핑 == 0 (예: ICSI 관련)
    ivf_case0_mask = ivf_mask & (train_data['시술유형매핑'] == 0)
    if ivf_case0_mask.sum() > 0:
        train_data.loc[ivf_case0_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case0_mask, '배아 이식 경과일'].fillna(0)
        icsi_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '미세주입된 난자 수', '미세주입에서 생성된 배아 수',
            '미세주입 배아 이식 수', '미세주입 후 저장된 배아 수', '해동된 배아 수'
        ]
        X_icsi = train_data.loc[ivf_case0_mask, icsi_features]
        y_icsi = train_data.loc[ivf_case0_mask, target]
        
        rf_model_icsi = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_icsi.fit(X_icsi, y_icsi)
        pred_rf_icsi = rf_model_icsi.predict(X_icsi)
        train_data.loc[ivf_case0_mask, '시술점수'] = pred_rf_icsi

    # ● IVF Case 1 : 시술유형매핑 == 1 (예: IVF 관련)
    ivf_case1_mask = ivf_mask & (train_data['시술유형매핑'] == 1)
    if ivf_case1_mask.sum() > 0:
        train_data.loc[ivf_case1_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case1_mask, '배아 이식 경과일'].fillna(0)
        ivf_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수',
            '이식된 배아 수', '미세주입에서 생성된 배아 수', '미세주입된 난자 수', '해동된 배아 수'
        ]
        X_ivf = train_data.loc[ivf_case1_mask, ivf_features]
        y_ivf = train_data.loc[ivf_case1_mask, target]
        
        rf_model_ivf = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_ivf.fit(X_ivf, y_ivf)
        pred_rf_ivf = rf_model_ivf.predict(X_ivf)
        train_data.loc[ivf_case1_mask, '시술점수'] = pred_rf_ivf

    # ● IVF Case 2 : 시술유형매핑 == 2 (예: Unknown 관련)
    ivf_case2_mask = ivf_mask & (train_data['시술유형매핑'] == 2)
    if ivf_case2_mask.sum() > 0:
        train_data.loc[ivf_case2_mask, '배아 이식 경과일_filled'] = train_data.loc[ivf_case2_mask, '배아 이식 경과일'].fillna(0)
        unknown_features = [
            '수집된 신선 난자 수', '총 생성 배아 수', '배아 이식 경과일_filled',
            '파트너 정자와 혼합된 난자 수', '혼합된 난자 수', '저장된 배아 수',
            '이식된 배아 수', '해동된 배아 수'
        ]
        X_unknown = train_data.loc[ivf_case2_mask, unknown_features]
        y_unknown = train_data.loc[ivf_case2_mask, target]
        
        rf_model_unknown = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model_unknown.fit(X_unknown, y_unknown)
        pred_rf_unknown = rf_model_unknown.predict(X_unknown)
        train_data.loc[ivf_case2_mask, '시술점수'] = pred_rf_unknown

print("Train_data에 시술점수 칼럼 추가 완료.")

# ================================
# 2. TEST_DATA 처리 및 예측 (학습한 모델 재사용)
# ================================
test_data['시술점수'] = np.nan

# ----- (1) DI 처리 (test_data) -----
di_mask_test = test_data['시술 유형'] == 'DI'
if di_mask_test.sum() > 0:
    test_data.loc[di_mask_test, '임신시도_filled'] = test_data.loc[di_mask_test, '임신 시도 또는 마지막 임신 경과 연수'].fillna(0)
    test_data.loc[di_mask_test, '클리닉시술_num'] = test_data.loc[di_mask_test, '클리닉 내 총 시술 횟수'].map(procedure_mapping)
    test_data.loc[di_mask_test, '불임원인'] = test_data.loc[di_mask_test, '부부 주 불임 원인'].astype(int)
    test_data.loc[di_mask_test, 'DI 시술'] = test_data.loc[di_mask_test, 'DI 시술 횟수'].map(procedure_mapping)
    
    X_DI_test = test_data.loc[di_mask_test, features_DI]
    pred_rf_di_test = rf_model_di.predict(X_DI_test)
    test_data.loc[di_mask_test, '시술점수'] = pred_rf_di_test

# ----- (2) IVF 처리 (test_data) -----
ivf_mask_test = test_data['시술 유형'] == 'IVF'
if ivf_mask_test.sum() > 0:
    # 상세 시술 유형("특정 시술 유형") 기준으로 매핑 (결측이면 'IVF:Unknown:Unknown:Unknown'으로 지정)
    test_data.loc[ivf_mask_test, '시술유형매핑'] = test_data.loc[ivf_mask_test, '특정 시술 유형']\
        .fillna('IVF:Unknown:Unknown:Unknown')\
        .map(ivf_mapping_dict)
    
    # ● IVF Case 0 (시술유형매핑 == 0)
    ivf_case0_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 0)
    if ivf_case0_mask_test.sum() > 0:
        test_data.loc[ivf_case0_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case0_mask_test, '배아 이식 경과일'].fillna(0)
        X_icsi_test = test_data.loc[ivf_case0_mask_test, icsi_features]
        pred_rf_icsi_test = rf_model_icsi.predict(X_icsi_test)
        test_data.loc[ivf_case0_mask_test, '시술점수'] = pred_rf_icsi_test

    # ● IVF Case 1 (시술유형매핑 == 1)
    ivf_case1_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 1)
    if ivf_case1_mask_test.sum() > 0:
        test_data.loc[ivf_case1_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case1_mask_test, '배아 이식 경과일'].fillna(0)
        X_ivf_test = test_data.loc[ivf_case1_mask_test, ivf_features]
        pred_rf_ivf_test = rf_model_ivf.predict(X_ivf_test)
        test_data.loc[ivf_case1_mask_test, '시술점수'] = pred_rf_ivf_test

    # ● IVF Case 2 (시술유형매핑 == 2)
    ivf_case2_mask_test = ivf_mask_test & (test_data['시술유형매핑'] == 2)
    if ivf_case2_mask_test.sum() > 0:
        test_data.loc[ivf_case2_mask_test, '배아 이식 경과일_filled'] = test_data.loc[ivf_case2_mask_test, '배아 이식 경과일'].fillna(0)
        X_unknown_test = test_data.loc[ivf_case2_mask_test, unknown_features]
        pred_rf_unknown_test = rf_model_unknown.predict(X_unknown_test)
        test_data.loc[ivf_case2_mask_test, '시술점수'] = pred_rf_unknown_test

print("Test_data에 시술점수 칼럼 추가 완료.")
```
